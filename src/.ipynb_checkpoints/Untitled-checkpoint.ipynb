```python
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df5[~df5['total_sqft'].apply(isfloat)].head(50)
```

```python
def convertRangeToFloat(x):
    tokens = x.split('-')
    if(len(tokens)==2):
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None   
```

```python
df5.shape
```

```python
df6 =df5.copy()
df6['total_sqft']  = df6['total_sqft'].apply(convertRangeToFloat)
```

```python
df6.isnull().sum()
```

```python
df7 = df6.dropna(subset=['total_sqft'])
```

```python
df8= df7.copy()
df8['price_per_sq_ft'] = df8['price']*100000/ df8['total_sqft']
```
